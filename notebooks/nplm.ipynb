## Neural Probabilistic Language Model

```python
words = [
    "Massimo", "Liza", "Miriam", "Giorgio", "Whisky", "Dodo",
    "James", "Emma", "Oliver", "Sophia", "Liam", "Ava", "Noah", "Isabella",
    "Ethan", "Mia", "Lucas", "Charlotte", "Mason", "Amelia", "Logan", "Harper",
    "Alexander", "Evelyn", "Benjamin", "Abigail", "Elijah", "Emily", "William",
    "Elizabeth", "Henry", "Sofia", "Jacob", "Avery", "Michael", "Ella",
    "Daniel", "Scarlett", "Matthew", "Grace", "Jackson", "Chloe", "Sebastian",
    "Victoria", "David", "Riley", "Joseph", "Aria", "Samuel", "Lily", "Carter",
    "Aubrey", "Owen", "Zoey", "Wyatt", "Penelope", "John", "Hannah", "Luke",
    "Layla", "Gabriel", "Nora", "Anthony", "Ellie", "Isaac", "Madeline", "Jack",
    "Hazel", "Jayden", "Violet", "Dylan", "Aurora", "Nathan", "Stella", "Caleb",
    "Lucy", "Ryan", "Anna", "Andrew", "Caroline", "Joshua", "Genesis",
    "Christopher", "Aaliyah", "Thomas", "Kennedy", "Charles", "Kinsley", "Josiah",
    "Allison", "Isaiah", "Sarah", "Levi", "Maria", "Mateo", "Iris", "Julian",
    "Naomi", "Aaron", "Eliana", "Elias", "Jade", "Connor", "Clara", "Cameron",
    "Elena",
]
```

```python
NUM_LETTERS = 26
NUM_SPECIAL = 1
NUM_TOKENS = NUM_LETTERS + NUM_SPECIAL

alphabet = list(bytes([list('a'.encode("utf-8"))[0] + i for i in range(NUM_LETTERS)]).decode('utf-8'))

stoi = {
    ".": 0
}
itos = {
    0: "."
}
for i, s in enumerate(alphabet):
    stoi[s] = i + 1
    itos[i + 1] = s

itos
```

```python
import torch

CONTEXT_SIZE = 3
X, Y = [], []

for w in words:
    w = w.lower()
    print(w)
    context = [0] * CONTEXT_SIZE
    for s in w + ".":
        i = stoi[s]
        X.append(context)
        Y.append(i)
        print(f"{[itos[i] for i in context]} -> {s}")
        context = context[1:] + [i]

X = torch.tensor(X)
Y = torch.tensor(Y)
```

```python
import torch

EMBEDDING_SIZE = 2

g = torch.Generator().manual_seed(1)

# 2-dimensional embedding for each character
C = torch.randn((NUM_TOKENS, EMBEDDING_SIZE), generator=g)
C
```

```python
NUM_HIDDED_NEURONS = 100

g = torch.Generator().manual_seed(1)

W1 = torch.randn((CONTEXT_SIZE * EMBEDDING_SIZE), NUM_HIDDED_NEURONS, generator=g)
b1 = torch.randn(NUM_HIDDED_NEURONS, generator=g)
```

```python
g = torch.Generator().manual_seed(1)

W2 = torch.randn(NUM_HIDDED_NEURONS, NUM_TOKENS, generator=g)
b2 = torch.randn(NUM_TOKENS, generator=g)
```

```python
logits = h @ W2 + b2
print(logits.shape)
logits
```

```python
counts = logits.exp()
prob = counts / counts.sum(1, keepdims = True)
prob
```

```python
log_likelyhood = prob[torch.arange(prob.shape[0]), Y].log().mean()
nll = -log_likelyhood
nll
```

```python
import torch.nn.functional as F

# efficient way of computing the nll out of the box
F.cross_entropy(logits, Y)
```

```python
BATCH_SIZE = 32

idx = torch.randint(0, X.shape[0], (BATCH_SIZE, ))
idx
```

```python
# training on minibatches
parameters = [C, W1, b1, W2, b2]

for p in parameters:
    p.requires_grad = True

def train(lre):
    for _ in range(100):
        idx = torch.randint(0, X.shape[0], (BATCH_SIZE, ))

        # forward
        emb = C[X[idx]] # sample a random subset (of size BATCH_SIZE) of the input
        h = torch.tanh(emb.view(-1, CONTEXT_SIZE * EMBEDDING_SIZE) @ W1 + b1)
        logits = h @ W2 + b2
        loss = F.cross_entropy(logits, Y[idx])

        # backward
        for p in parameters:
            p.grad = None
        loss.backward()

        # update
        for p in parameters:
            p.data += -lre * p.grad
    
    return loss

train(0.1)
```

```python
# tuning the learning rate hyperparameter
lres = torch.linspace(-3, 0, 1000)
lres = 10**lres
lres
```

```python
losses = []
for lre in lres:
    loss = train(lre)
    print(loss.item())
    losses.append(loss)
```

```python
from matplotlib import pyplot as plt

plt.plot(lres, [loss.detach().numpy() for loss in losses])
```
