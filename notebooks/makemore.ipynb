```python
import torch
```

```python
NUM_LETTERS = 26
NUM_SPECIAL = 1
NUM_TOKENS = NUM_LETTERS + NUM_SPECIAL

N = torch.zeros((NUM_TOKENS, NUM_TOKENS))
N
```

```python

alphabet = list(bytes([list('a'.encode("utf-8"))[0] + i for i in range(NUM_LETTERS)]).decode('utf-8'))
alphabet
```

```python
stoi = {c: i + 1 for i, c in enumerate(alphabet)}
stoi[SPECIAL_TOKEN] = 0
stoi
```

```python
itos = {i: s for s, i in stoi.items()}
itos
```

```python
for w in words:
    chars = [SPECIAL_TOKEN] + list(w.lower()) + [SPECIAL_TOKEN]
    for c1, c2 in zip(chars, chars[1:]):
        idx = stoi[c1]
        idy = stoi[c2]
        N[idx][idy] += 1

N
```

```python
p = N[0].float() / sum(N[0])
p
```

```python
g = torch.Generator().manual_seed(214783647)
idx = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
itos[idx]
```

```python
g = torch.Generator().manual_seed(214783647)

idx = 0
word = ""
while True:
    p = N[idx].float() / sum(N[idx])
    idx = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
    if idx == 0:
        break
    word += itos[idx]

word
```
