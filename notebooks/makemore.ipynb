```python
class NLP(MLP):
    
    def get_probability(self, x, y):
        # one hot encoding
        xenc = [1.0 if i == x else 0.0 for i in range(NUM_TOKENS)]
        # exponentiate the outputs so they are all > 0
        ypreds = [ypred.exp() for ypred in self(xenc)]
        # normalize them so that they represent a probability distribution
        probs = [ypred / sum(ypreds) for ypred in ypreds]
        # return the probability of the expected output
        # we would like this probability to be high
        return probs[y]

    def train(self, xs, ys, iter, step):
        for _ in range(iter):
            # forward
            probs = [self.get_probability(x, y) for x, y in zip(xs, ys)]
            # likelyhood = product of probabilities
            # goal: maximize the likelyhood
            # to make sure the likelyhood is not too small we take the logarithm
            # log(p1 * p2) = log(p1) + log(p2)
            # for convenience, we consider the average of the logprobs instead of the sum
            # log of probabilities (<= 1) is <= 0
            # consider the negative log likelyhood as the loss to be minimized
            logprobs = [prob.log() for prob in probs]
            log_likelyhood = sum(logprobs) / len(logprobs)
            loss = -log_likelyhood
            print(f"loss {loss.data}")
            loss._label = "loss"
            loss.grad = 1
            self.zero_grad()
            # backward
            loss.backward()
            # update
            for p in self.parameters():
                p.data += p.grad * (-step)
```

```python
nlp = NLP(NUM_TOKENS, [NUM_TOKENS])
```

```python
import torch.nn.functional as F
```

```python
xs = torch.tensor(xs)
ys = torch.tensor(ys)
```

```python
g = torch.Generator().manual_seed(1234567890)
W = torch.randn((NUM_TOKENS, NUM_TOKENS), generator=g, requires_grad=True)
```

```python
xenc = F.one_hot(xs, num_classes=NUM_TOKENS).float()
# predict log-counts
logits = xenc @ W
# softmax
counts = logits.exp()
# probabilities for the next character
probs = counts / counts.sum(1, keepdims=True)
```

```python
# average negative log likelyhood
loss = -probs[torch.arange(len(xs)), ys].log().mean()
loss
```

```python
# set gradients to zero
W.grad = None
loss.backward()
```

```python
W.grad
```

```python
W.data += -0.1 * W.grad
```

```python
for k in range(100):
    # forward
    xenc = F.one_hot(xs, num_classes=NUM_TOKENS).float()
    logits = xenc @ W
    counts = logits.exp()
    probs = counts / counts.sum(1, keepdims=True)
    loss = -probs[torch.arange(len(xs)), ys].log().mean()
    print(loss.data)

    # backward
    W.grad = None
    loss.backward()

    # update
    W.data += -50 * W.grad
```
